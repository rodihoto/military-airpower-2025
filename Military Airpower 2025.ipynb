```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Data
data = {
    'Country': ['United States', 'Russia', 'China', 'India', 'South Korea',
                'Japan', 'Pakistan', 'Egypt', 'Turkey', 'France'],
    'Total_Aircraft': [13043, 4292, 3309, 2229, 1592, 1443, 1399, 1093, 1083, 976]
}
df = pd.DataFrame(data)

# 📊 Bar Chart
plt.figure(figsize=(12, 7))
bars = plt.bar(df['Country'], df['Total_Aircraft'], color='lightblue')
plt.title('Top 10 Countries by Total Military Aircraft (2025)', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Total Aircraft')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:,}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), textcoords="offset points", ha='center', fontsize=9)
plt.tight_layout()
plt.savefig("top_10_aircraft_2025.png", dpi=300)
plt.show()

# 🥧 Pie Chart
plt.figure(figsize=(10, 8))
plt.pie(df['Total_Aircraft'], labels=df['Country'], autopct='%1.1f%%',
        startangle=140, textprops={'fontsize': 10})
plt.title('Global Distribution of Military Aircraft (Top 10 Countries, 2025)', fontsize=14)
plt.tight_layout()
plt.savefig("top_10_aircraft_pie_2025.png", dpi=300)
plt.show()

# 🔥 Seaborn Heatmap Table
plt.figure(figsize=(10, 6))
table_data = df.sort_values('Total_Aircraft', ascending=False).set_index('Country')
sns.heatmap(table_data.T, annot=True, fmt=',', cmap='Blues', cbar=False,
            linewidths=0.5, linecolor='grey', annot_kws={"size": 12})
plt.title('Total Military Aircraft by Country (2025)', fontsize=14)
plt.yticks(rotation=0)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("aircraft_table_heatmap_2025.png", dpi=300)
plt.show()

# 📋 Clean Table with Matplotlib
fig, ax = plt.subplots(figsize=(10, 5))
ax.axis('off')
table = plt.table(cellText=df.values,
                  colLabels=df.columns,
                  cellLoc='center',
                  loc='center',
                  colColours=['lightblue'] * 4)
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.5)
plt.title("Top 10 Countries by Total Military Aircraft (2025)", fontsize=14)
plt.savefig("aircraft_table_clean_2025.png", dpi=300)
plt.show()

# 📈 Plotly Interactive Bar Chart
fig = px.bar(df.sort_values("Total_Aircraft", ascending=False),
             x="Country",
             y="Total_Aircraft",
             title="Top 10 Countries by Total Military Aircraft (2025)",
             text="Total_Aircraft",
             labels={"Total_Aircraft": "Total Aircraft"},
             template="plotly_white")
fig.update_traces(texttemplate='%{text:,}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.write_html("aircraft_top10_plotly_2025.html")
```
